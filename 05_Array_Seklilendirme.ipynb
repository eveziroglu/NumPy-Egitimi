# Dizi Şekillendirme, Düzleştirme ve Dönüştürme

Bu notebookta, NumPy kütüphanesindeki dizilerin şekillendirilmesi, düzleştirilmesi ve boyutlarının dönüştürülmesi konularını ele alacağız.

## 1. Dizi Şekillendirme

Dizileri yeniden şekillendirmek, dizinin boyutunu değiştirmek anlamına gelir. Örneğin, bir dizi biçimini (2, 3) olarak değiştirebiliriz.

```python
import numpy as np

dizi = np.array([1, 2, 3, 4, 5, 6])
dizi_sekillendirilmis = dizi.reshape(2, 3)
print(dizi_sekillendirilmis)
```

## 2. Dizi Düzleştirme

Bir dizi düzleştirmek, çok boyutlu diziyi tek bir boyutlu dizi haline getirmek demektir. NumPy'de `flatten()` yöntemi kullanılır.

```python
dizi_duzlestirilmis = dizi_sekillendirilmis.flatten()
print(dizi_duzlestirilmis)
```

## 3. Dizi Boyutlarını Dönüştürme

Boyutları dönüştürmek için `np.newaxis` veya `reshape` yöntemini kullanabiliriz.

```python
dizi_yeni_boyut = dizi[np.newaxis, :]
print(dizi_yeni_boyut)
```
